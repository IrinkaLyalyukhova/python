```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Распределение Лапласа 

https://ru.wikipedia.org/wiki/Распределение_Лапласа

```python
loc, scale = 1., 1.2

laplace = sts.laplace(loc, scale)
samples = laplace.rvs(1000)

x = np.linspace(-10., 10., 1000)
den = laplace.pdf(x)

plt.plot(x, den, label='PDF')
plt.hist(samples, density=True, label='Histogramm')
plt.xlabel('$x$')
plt.legend()
```

Находим математическое ожидание и дисперсию, строим аппроксимацию

```python
mean = laplace.mean()
var = laplace.var()

def approximation(norm_menan, norm_var,size):  
  n = laplace.rvs([1000,size])
  n_mean = n.mean(axis=1)
  plt.figure()
  plt.hist(n_mean, bins=30, density=True)
  nrmApprox = sts.norm(norm_menan, pow(norm_var/size, 0.5))
  
  locDomain = np.linspace(-10., 10., 1000)
  normPdf = nrmApprox.pdf(locDomain)
  plt.plot(locDomain, normPdf)
  
   
approximation(mean, var, 5)
approximation(mean, var, 10)
approximation(mean, var, 50)
approximation(mean, var, 100)
```
